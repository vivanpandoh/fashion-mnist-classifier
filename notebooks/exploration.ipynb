# Fashion-MNIST Exploration & Demo

This notebook walks through loading the data, visualizing samples, training a small run, and inspecting predictions. For full training, use `python -m src.train` from the project root instead — this notebook is for exploration and demos.

```python
import sys
sys.path.append('..')

import numpy as np
import matplotlib.pyplot as plt

from src.data import load_data, make_train_val_split, make_datasets, CLASS_NAMES
from src.model import build_model, compile_model
```

```python
(x_train, y_train), (x_test, y_test) = load_data()
print('Train:', x_train.shape, 'Test:', x_test.shape)
```

## Sample images per class

```python
fig, axes = plt.subplots(2, 5, figsize=(12, 5))
for class_idx, ax in enumerate(axes.flat):
    idx = np.where(y_train == class_idx)[0][0]
    ax.imshow(x_train[idx].squeeze(), cmap='gray')
    ax.set_title(CLASS_NAMES[class_idx])
    ax.axis('off')
plt.tight_layout()
plt.show()
```

## Class distribution

```python
counts = np.bincount(y_train)
plt.figure(figsize=(9, 4))
plt.bar(CLASS_NAMES, counts)
plt.xticks(rotation=45, ha='right')
plt.ylabel('Count')
plt.title('Training set class distribution')
plt.tight_layout()
plt.show()
```

## Quick training demo (few epochs, for illustration only)

For the real, full run that reproduces the reported metrics, use the CLI: `python -m src.train --epochs 40 --batch-size 64`.

```python
(x_tr, y_tr), (x_val, y_val) = make_train_val_split(x_train, y_train)
train_ds, val_ds, test_ds = make_datasets(x_tr, y_tr, x_val, y_val, x_test, y_test, batch_size=64)

model = build_model()
model = compile_model(model)
history = model.fit(train_ds, validation_data=val_ds, epochs=3)
```

```python
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='train')
plt.plot(history.history['val_accuracy'], label='val')
plt.title('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='val')
plt.title('Loss')
plt.legend()
plt.tight_layout()
plt.show()
```

## Inspect a few predictions

```python
preds = model.predict(x_test[:10])
pred_labels = np.argmax(preds, axis=1)

fig, axes = plt.subplots(2, 5, figsize=(12, 5))
for i, ax in enumerate(axes.flat):
    ax.imshow(x_test[i].squeeze(), cmap='gray')
    correct = pred_labels[i] == y_test[i]
    color = 'green' if correct else 'red'
    ax.set_title(f'{CLASS_NAMES[pred_labels[i]]}', color=color, fontsize=9)
    ax.axis('off')
plt.tight_layout()
plt.show()
```
